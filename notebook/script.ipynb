```python
def is_valid_uuid(val):
    """

    :param val:
    :return:
    """
    try:
        return str(uuid.UUID(str(val)))
    except Exception:
        if str(val).isdigit() and len(str(val)) == 19:
            return str(val)
        else:
            return np.nan
```

```python
def post_code_validation(val):
    """

    :param val:
    :return:
    """
    try:
        if not pd.isna(val):
            # check for a 3 digit and 4 digit numbers. If 3 digit add leading 0.
            if re.search(r"(?<!\d)\d{3,4}(?!\d)", val):
                return re.findall(r"(?<!\d)\d{3,4}(?!\d)", val)[0].zfill(4)
            else:
                # check for a 5 digit numbers starting with 0.
                if re.search(r"(?<!\d)\d{5}(?!\d)", val):
                    return re.findall(r"(?<!\d)\d{5}(?!\d)", val)[0][1:]
                else:
                    # for everything else return nan value.
                    return np.nan
        else:
            return np.nan
    except Exception:
        raise Exception
```

```python
def gender_validation(val):
    """

    :param val:
    :return:
    """
    try:
        if str(val).strip()[0].upper() == 'M':
            return 'M'
        elif str(val).strip()[0].upper() == 'F':
            return 'F'
        else:
            # ['UNK', '\x0bOther', 'Unknown', 'Other', 'gender']
            return np.nan
    except Exception:
        raise Exception
```

```python
def birthdate_validation(val):
    try:
        return pd.to_datetime(val)
    except Exception:
        return np.nan
```

```python
def is_valid_int(value):
        """

        :param value:
        :return:
        """
        try:
            if len(str(value)) <= 32:
                if isinstance(value, float):
                    return int(value)
                elif isinstance(value, str):
                    return int(float(value))
                else:
                    return np.nan
            else:
                return np.nan
        except Exception:
            return np.nan
```

```python
unique_df = df.drop_duplicates()
unique_count = unique_df.shape[0]
unique_count
```

```python
unique_df.loc[:, 'transformed_donor_id'] = unique_df['donor_id'].map(is_valid_uuid)
```

```python
unique_df.loc[:, 'transformed_postcode'] = unique_df['postcode'].map(post_code_validation)
```

```python
unique_df.loc[:, 'transformed_gender'] = unique_df['gender'].map(gender_validation)
```

```python
unique_df.loc[:, 'transformed_birth_date'] = unique_df['birth_date'].map(birthdate_validation)
```

```python
unique_df.loc[:, 'transformed_donor_type'] = unique_df['donor_type'].map(is_valid_int)
```
